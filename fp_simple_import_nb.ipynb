# Get Input Data

```
df=np.asarray(df)
```

**Shuffle** the data so it's no longer in chronological order

```
np.random.shuffle(df)
hda_odds=df[:,[10,11,12]]
results=df[:,[17,18,19]]
```

# Reshape

Reshape

```
results=results.reshape(results.shape[0],1,1,3)
hda_odds=hda_odds.reshape(hda_odds.shape[0],1,1,3)
```

# Make Input Data float32 and seperate

```
hda_odds=hda_odds.astype('float32')
```

Usually approx 80% of the data is used as training data and 20% of the data is used as test data

```
p=0.9
number_dp=hda_odds.shape[0];
last_train= math.ceil(p*number_dp)
odds_train=hda_odds[0:last_train,:]
odds_test=hda_odds[(last_train+1):number_dp,:]
results_train=results[0:last_train,:]
results_test=results[(last_train+1):(number_dp),:]
```

# Define Model Architecture

```
model=Sequential()
model.add(Dense(9,input_shape=(1,1,3), activation='relu')) # Adds a densely connected NN layer
model.add(Dense(3,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

# Compile Model

Categorial crossentropy loss is a loss algorithm for when there are multiple categories to sort each data point into, in this case  home win, draw or away win. The optimizer is a feature that makes the model work better, adam is the default.

```
opt = optimizers.SGD(lr=0.1,momentum=0.004)

model.compile(loss='categorical_crossentropy',optimizer=opt,metrics=['accuracy']) 
# Want to know the accuracy of the model so specify it as a metric
plot_model(model,to_file='model.png')

```

# Fit Model from Training Data


The number of epochs are the number of phases of training. The shuffle attribute determines whether the training data is shuffled between epochs

```


model.fit(odds_train,results_train,batch_size=100,epochs=200,verbose=1,shuffle=1)
```

# Evaluate Model

```
[loss_value,accuracy]=model.evaluate(odds_test,results_test,verbose=1)
print("The loss value for the model is "+ str(loss_value))
print("The accuracy value for the model is "+str(accuracy))
```

# Make Prediction

Getting the model to predict the probability of each outcome for a game with  hda odds: 10, 5, 1.85 (away win likely)

```
h_favoured=np.asarray([1.85, 5, 10]).reshape(1,1,1,3)
a_favoured=np.asarray([10,5,1.85]).reshape(1,1,1,3)

prediction=model.predict(a_favoured)
print(str(prediction))
```
